```python
player_1 = input("player 1, select your choice of (rock, paper, scissors): ")
player_2 = input("player 2, select your choice of (rock, paper, scissors): ")
```

```python
print(f"\nplayer_1 chose: {player_1}")
print(f"player_2 chose: {player_2}\n")
```

```python
if player_1 == player_2: 
    print("we have a tie!, play again!") 
elif (player_1 == 'rock' and player_2 == 'scissors') or \
    (player_1 == 'scissors' and player_2 == 'paper') or \
    (player_1 == 'paper' and player_2 == 'rock'):
    print ("Horray, player 1! You win!")
else: 
    print("Horray, player 2! You win!")
```
